```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```python
import glob
files = glob.glob("*.csv")
```

```python
data = [pd.read_csv(file, encoding = 'ISO-8859-1') for file in files]
```

```python
def clean_data(df):
    # 'Remove existing header with unnames'
    df=df.rename(columns=df.iloc[0]).drop(df.index[0])
    # 'Renames country in column 1 as country'
    df.columns = df.columns.fillna('Country')
    df_2015 = df[df.Year == '2015']
    # 'Reformat Value from string to flue for computation'
    df_2015.Value.replace(to_replace = ',', value= '_', regex = True, inplace = True) 
    df_2015.Value = df_2015.Value.astype(dtype = 'float64')
    # 'Return dataframe with Series entries as columns'
    df_pop = pd.pivot_table(df_2015, columns = 'Series', index =['Country', 'Year'])
    return df_pop
```

```python
clean_frames = []
for df in data:
    clean_frames.append(clean_data(df))
```
